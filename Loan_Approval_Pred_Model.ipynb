```python
#pandas is mainly use for reading data
import pandas as pd
```

```python
#reading data set with panda
# You can download dataset from = https://drive.google.com/file/d/15gEtHFxPtkUeGqqDkYzYmr5rDfZVLux1/view?usp=drive_link
data = pd.read_csv('loan_approval_dataset.csv')
```

```python
#droping irrelevant column
data.drop(columns = ['loan_id'], inplace=True)
```

```python
#Checking columns name after dropping the loan_id
data.columns
```

```python
#removing irrelevant spaces from column names
data.columns = data.columns.str.strip()
```

```python
#checking if spaces get remove
data.columns
```

```python
#checking data
data
```

```python
#checking columns name to make same type column as one and remove it
data.columns
```

```python
#making new column asset for all the assest type columns
data['Assets'] = data.residential_assets_value + data.commercial_assets_value+data.luxury_assets_value + data.bank_asset_value
```

```python
#dropping all the asset type columns
data.drop(columns = ['residential_assets_value','commercial_assets_value','luxury_assets_value','bank_asset_value'], inplace = True)
```

```python
#chcking if columns get remove and asset column get added
data
```

```python
#checking if there is any null value in the data
data.isnull().sum()
```

```python
#checking data after checking null values
data
```

```python
#checking all the unique value in education column (unqiue is used to print/count all repeated and non repeated value once only)
data.education.unique()
```

```python
#making function to remove the irrelevant spaces from data
def clean_data(st):
    st = st.strip()
    return st
```

```python
#checking if funcation working properly
clean_data(' Graduate')
```

```python
#removing irrelevant spaces from education column by giving it to function clean_data
data.education = data.education.apply(clean_data)
```

```python
#checking all the unique value in education column after removing spaces
data.education.unique()
```

```python
#replacing Graduate to 1 and Not Graduate to 0
data['education'] = data['education'].replace(['Graduate', 'Not Graduate'],[1,0])
```

```python
#checking data after replacing
data
```

```python
#checking all the unique value in self_employed column
data.self_employed.unique()
```

```python
#removing irrelevant spaces from self_employed column by giving it to function clean_data
data.self_employed = data.self_employed.apply(clean_data)
```

```python
#checking all the unique value in self_employed column after removing spaces
data.self_employed.unique()
```

```python
#replacing No to 0 and Yes to 1
data.self_employed = data.self_employed.replace(['No', 'Yes'],[0,1])
```

```python
#checking data after replacing
data
```

```python
#checking all the unique value in loan_status column
data.loan_status.unique()
```

```python
#removing irrelevant spaces from loan_status column by giving it to function clean_data
data.loan_status = data.loan_status.apply(clean_data)
```

```python
#checking all the unique value in loan_status column after removing spaces
data.loan_status.unique()
```

```python
##replacing Approved to 1 and No to 0
data.loan_status = data.loan_status.replace(['Approved', 'Rejected'],[1,0])
```

```python
#checking data after replacing
data
```

```python
#importing train_test_split method
from sklearn.model_selection import train_test_split
```

```python
#dividing data into 2 part input data [all the colums excepts loan_status] and output data [loan_status]
input_data = data.drop(columns=['loan_status'])
output_data = data['loan_status']
```

```python
#checking input data
input_data
```

```python
#checking output data
output_data
```

```python
#dividing into train and test part (train is used for training the model and test is use for check whether it can predict right or wrong)
#taking 4 variables because train_test_split return 4 outputs
#train_input,test_input,train_output,test_output = train_test_split(input data, output data, train_size = 0.2{which means 20% data is in test and 80% in train}, random_state = 1(this is use because with this every time program will take same data as input data and same as output data for both train and test))

x_train,x_test,y_train,y_test = train_test_split(input_data,output_data, test_size=0.2, random_state=1)
```

```python
#checking the size of all the 4 variables
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
#calling standardscaler for standardization (it is important to standardize data because it will lead to less retundancy, differnce between column values reduces(1,2400000,3), less error )
from sklearn.preprocessing import StandardScaler
```

```python
#making instance of StandardScaler class
scaler = StandardScaler()
```

```python
#standardizing input data for train
x_train_scaled = scaler.fit_transform(x_train)
```

```python
#standardizing input data for test
x_test_scaled = scaler.transform(x_test)

#Output data is in 0 and 1 only so we don't need to standardize it
```

```python
#importing model
from sklearn.linear_model import LogisticRegression
```

```python
#making instance of class logiticRegression
model = LogisticRegression()
```

```python
#fiting input_train and output_train on model or making model learn from input_train and output_train
model.fit(x_train_scaled, y_train)
```

```python
#checking its accuracy with input_test and output_test
model.score(x_test_scaled, y_test)
```

```python
#storing input data into variable
pred_data = pd.DataFrame([['2','1','0','9600000','29900000','12','778','50700000']],columns=['no_of_dependents','education','self_employed','income_annum','loan_amount','loan_term','cibil_score','Assets'])
```

```python
#standardizing input data
pred_data = scaler.transform(pred_data)
```

```python
#prediciting wether loan approved or rejected
result = model.predict(pred_data)
if result==1:
    print('Approved')
else:
    print('Rejected')
```
